```python
df = spark.table("workspace.default.people_1000")

display(df)
df.printSchema()
```

```python

df_clean = df.toDF(
    *[col.replace(" ", "_").replace("-", "_").replace("/", "_") for col in df.columns]
)

df_clean.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("workspace.default.people_1000_delta")
```

```python
df = spark.table("workspace.default.people_1000_delta")
display(df)
```

```python
df_filtered = df.filter("sex == 'Female'")

display(df_filtered)  
```
